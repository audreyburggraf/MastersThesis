```python
import numpy as np

# Set up matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.patches import Circle
from matplotlib.colors import LogNorm

from astropy.coordinates import SkyCoord
import astropy.units as u
from astropy.wcs import WCS
from astropy.io import fits
from astropy.table import Table
from astropy.visualization import astropy_mpl_style

```

```python
# Font sizes
title_fs = 25
axis_label_fs = 20
axis_num_fs = 20
legend_fs = 20
cbar_fs = 20
text_fs = 28
```

```python
file = 'perseus+WholeMap_herschelONLY+updatedCCs_beta2_mu2_8_columnDensityMap+grid.fits'
```

```python
hdu_list = fits.open(file, memmap=True)
```

```python
header = hdu_list[0].header
```

```python
data = hdu_list[0].data  # Image data to plot
```

```python
# Set the plotting style for better visuals
plt.style.use(astropy_mpl_style)

# Create the WCS object
wcs = WCS(header)

# Create the plot with WCS projection
fig, ax = plt.subplots(subplot_kw={'projection': wcs})
ax.imshow(data, origin='lower', cmap='gray')  # Plot the image data

# Set RA and Dec labels in the required formats
ax.coords[0].set_major_formatter('hh:mm:ss')  # RA in hours, minutes, seconds
ax.coords[1].set_major_formatter('dd:mm:ss')  # Dec in degrees, arcminutes, arcseconds

# Label the axes
ax.set_xlabel('Right Ascension (hh:mm:ss)')
ax.set_ylabel('Declination (dd:mm:ss)')

plt.show()
```

```python
# Define plot size
fig, ax = plt.subplots(figsize=(14, 12), subplot_kw={'projection': wcs})

# Overlay data with a logarithmic color scale using the 'viridis' colormap
# Adjust vmin and vmax to control the range for log normalization
im = ax.imshow(data, 
               cmap='viridis', 
               vmin=0.5e21, vmax=2.0e23, 
               norm=LogNorm())

# Add a colorbar
cbar = plt.colorbar(im, ax=ax)
cbar.set_label(r'Column Density (cm$^{-2}$)', fontsize=14)

# Set the title and labels
ax.set_title('Column Density Map of Perseus', fontsize=16, fontweight='bold')
ax.set_xlabel('RA (hh:mm:ss)', fontsize=14)
ax.set_ylabel('Dec (dd:mm:ss)', fontsize=14)

# Format RA and Dec axes in hh:mm:ss and dd:mm:ss
ax.coords[0].set_major_formatter('hh:mm:ss')  # RA
ax.coords[1].set_major_formatter('dd:mm:ss')  # Dec

# Customize tick parameters
ax.tick_params(labelsize=12, which='major', length=7)
ax.tick_params(which='minor', length=4)
cbar.ax.tick_params(labelsize=12, which='major', length=7)
cbar.ax.tick_params(which='minor', length=4)

# Add a circular beam to the plot
beam_circle = Circle(
    (0.5, 0.5),  # Adjust this position based on your data's coordinates
    radius=0.05,  # Adjust beam radius based on actual data needs
    edgecolor='black',
    facecolor='blue',
    alpha=1,
    label='Beam Size (x30)'
)
ax.add_patch(beam_circle)

# Add text label in the top left corner
ax.text(0.05, 0.95, 'Perseus', transform=ax.transAxes, fontsize=14, ha='left', va='top')

# Add an example line and label for distance scale in AU
x_start_deg, y_pos_deg = 0.9, 0.1  # Adjust based on data
length_deg = 0.05  # Example length
ax.plot([x_start_deg, x_start_deg + length_deg], [y_pos_deg, y_pos_deg], color='black', linewidth=3)
ax.text(x_start_deg + length_deg / 2, y_pos_deg - 0.05, f'{int(length_deg * 206265)} AU', 
        fontsize=12, ha='center', va='top')

# Flip the y-axis if needed
ax.invert_yaxis()

# Add a legend
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), fontsize=12, ncol=1)

# Tight layout for better spacing
plt.tight_layout()
plt.show()
```
