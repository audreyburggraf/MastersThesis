```python
lambda_band1_micron  = 7.0  * 1000  # mm to microns
lambda_band3_micron  = 3.1  * 1000  # mm to microns
lambda_band7_micron  = 0.87 * 1000  # mm to microns
lambda_band10_micron = 0.34 * 1000  # mm to microns
```

```python
lambda_targets = [lambda_band1_micron, lambda_band3_micron, lambda_band7_micron, lambda_band10_micron]
```

```python
m_re_interp = []
m_im_interp = []


for i in range(4):
    m_re = np.interp(lambda_targets[i], df_water_ice["wavelength (microns)"], df_water_ice["m_re"])
    m_im = np.interp(lambda_targets[i], df_water_ice["wavelength (microns)"], df_water_ice["m_im"])
    
    m_re_interp.append(m_re)
    m_im_interp.append(m_im)
    
```

```python
m_complex_interp = [n + 1j * k for n, k in zip(m_re_interp, m_im_interp)]
```
