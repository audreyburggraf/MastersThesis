## Importing Stuff

```python
import numpy as np

# matplotlib
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

import seaborn as sns

from custom_colormap import soft_colormap_v2
```

```python
# Font sizes
title_fs = 25
axis_label_fs = 20
axis_num_fs = 20
legend_title_fs = 20
legend_text_fs = 15
cbar_fs = 20
text_fs = 28
```

I asked ChatGPT to make me some mock data that is circular, has minor and major axes that I can set, and thre is an easy-to-see trend on each axis to make it easy to see if I am slicing the data correctly. Here is what it game me:

```python
# Parameters
num_pixels = 100
x, y = np.meshgrid(np.arange(num_pixels), np.arange(num_pixels))
center_x, center_y = num_pixels // 2, num_pixels // 2  # Center of the data
major_angle_rad = np.radians(45)  # Major axis at 45 degrees
minor_angle_rad = np.radians(135)  # Minor axis at 135 degrees
radius = np.sqrt((x - center_x)**2 + (y - center_y)**2)  # Radial distance from the center

# Create synthetic data with circular symmetry and variations along axes
radial_pattern = np.cos(radius / 10)  # Circular oscillations

# Gaussian weighting for major axis (rotated at 45 degrees)
major_component = np.exp(-0.5 * ((np.cos(major_angle_rad) * (x - center_x) + 
                                   np.sin(major_angle_rad) * (y - center_y))**2) / (2 * 20**2))

# Gaussian weighting for minor axis (rotated at 135 degrees)
minor_component = np.exp(-0.5 * ((np.cos(minor_angle_rad) * (x - center_x) + 
                                   np.sin(minor_angle_rad) * (y - center_y))**2) / (2 * 15**2))

# Combine components with radial pattern
data = radial_pattern * (50 * major_component + 30 * minor_component)

# Normalize the data to range [0, 100]
data -= np.min(data)
data /= np.max(data)
data *= 100

# Round the data to the nearest multiple of 10
data = np.round(data / 10) * 10
```

```python
line_length = np.sqrt(2 * num_pixels**2)

# Generate points along the major axis
major_x = center_x + np.linspace(-line_length, line_length, num_pixels) * np.cos(major_angle_rad)
major_y = center_y + np.linspace(-line_length, line_length, num_pixels) * np.sin(major_angle_rad)

# Generate points along the minor axis
minor_x = center_x + np.linspace(-line_length, line_length, num_pixels) * np.cos(minor_angle_rad)
minor_y = center_y + np.linspace(-line_length, line_length, num_pixels) * np.sin(minor_angle_rad)

# Extract data along the major axis using bilinear interpolation
major_data = []
valid_major_x = []  # To store valid x positions
valid_major_y = []  # To store valid y positions

for i in range(len(major_x)):
    xi, yi = int(round(major_x[i])), int(round(major_y[i]))
    if 0 <= xi < num_pixels and 0 <= yi < num_pixels:
        major_data.append(data[yi, xi])
        valid_major_x.append(xi)  # Store valid x position
        valid_major_y.append(yi)  # Store valid y position

# Extract data along the minor axis using bilinear interpolation
minor_data = []
valid_minor_x = []  # To store valid x positions
valid_minor_y = []  # To store valid y positions

for i in range(len(minor_x)):
    xi, yi = int(round(minor_x[i])), int(round(minor_y[i]))
    if 0 <= xi < num_pixels and 0 <= yi < num_pixels:
        minor_data.append(data[yi, xi])
        valid_minor_x.append(xi)  # Store valid x position
        valid_minor_y.append(yi)  # Store valid y position
```

```python
# Create a custom color palette with 11 very distinct colors
custom_cmap = sns.color_palette([
    "#e81416",  # Red
    "#ffa500",  # Orange
    "#faeb36",  # Yellow
    "#79c314",  # Green
    "#487de7",  # Blue
    "#8a2be2",  # Purple
    "#ff1493",  # Deep Pink
    "#ff69b4",  # Hot Pink
    "#000000",  # Black
    "#808080",  # Grey
    "#ffffff"   # White (optional)
])

# Convert the Seaborn palette to a Matplotlib colormap and add alpha channel (alpha = 1 by default)
custom_cmap = [(r, g, b, 1) for r, g, b in custom_cmap]

# Now you can change the alpha value dynamically
alpha_value = 0.5
custom_cmap = ListedColormap([(r, g, b, alpha_value) for r, g, b, _ in custom_cmap])

custom_cmap
```

```python
# Plot the 2D data grid
fig, ax = plt.subplots(figsize=(14, 12))  # Same size as your previous plot

# Plot the image with the custom colormap
im = ax.imshow(data, cmap= custom_cmap,
              origin='lower', extent=[0, num_pixels, 0, num_pixels])

# Colorbar with custom font sizes
# Colorbar with custom font sizes
cbar = plt.colorbar(im, ax=ax, ticks=np.linspace(0, 100, 11))
cbar.set_label('Data Value', fontsize=cbar_fs)
cbar.ax.tick_params(labelsize=axis_num_fs, which='major', length=7, direction="in")
cbar.ax.tick_params(which='minor', length=4, direction="in")

# Remove title and adjust axis labels with custom font sizes
ax.set_xlabel('X (pixels)', fontsize= axis_label_fs)
ax.set_ylabel('Y (pixels)', fontsize=axis_label_fs)

# Set axis limits
ax.set_xlim(0, num_pixels)
ax.set_ylim(0, num_pixels)

# Add minor and major axes
plt.plot(major_x, major_y, 'r-', lw = 5, label='Major Axis')
plt.plot(minor_x, minor_y, 'b-', lw = 5, label='Minor Axis')

# Add centre point
ax.scatter(center_x, 
           center_y,
           color='darkblue', 
           marker='*', 
           s=2000,
           label = 'Centre Point')

# Add a legend
plt.legend(fontsize = legend_text_fs)



# Adjust ticks
ax.minorticks_on()
ax.tick_params(axis="x", which="major", direction="in", bottom=True, top=True, length=7, labelsize= axis_num_fs)
ax.tick_params(axis="y", which="major", direction="in", left=True, right=True, length=7, labelsize=axis_num_fs)

ax.axvline(x=90, color="white", linestyle="-", linewidth=5)
ax.axhline(y=90, color="white", linestyle="-", linewidth=5)


# Show the plot
plt.show()
```
