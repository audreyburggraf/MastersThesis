In this notebook I am going to figure out how I can save the data I need for Gaussian

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Font sizes
title_fs = 25
axis_label_fs = 20
axis_num_fs = 20
legend_title_fs = 20
legend_text_fs = 15
cbar_fs = 20
text_fs = 28
```

```python
a_values = [0, 1, 2, 3]
b_values = [4, 5, 6, 7]
c_values = [8, 9, 10, 11]
```

```python
def d_function(a, b, c):
    
    return a**2 + b**2 + c**2
```

```python
# Dictionary to store results
results_2d = []

c_const = 1

for a in a_values: 
    for b in b_values: 
            
        d = d_function(a, b, c_const)

        results_2d.append((a, b, d))
```

```python
results_2d
```

Now I want to extract the data from the results

```python
# Convert results into structured 2D arrays
a_list, b_list, d_list = zip(*results_2d)  # Unpack into separate lists

# Reshape into 2D grids (len(a_values) rows, len(b_values) columns)
A = np.array(a_list).reshape(len(a_values), len(b_values))
B = np.array(b_list).reshape(len(a_values), len(b_values))
D = np.array(d_list).reshape(len(a_values), len(b_values))
```

Now lets plot it

```python
# Plotting the result
fig, ax = plt.subplots(figsize=(10, 8))

# Contour plot
contour = ax.contourf(A, B, D, 50, cmap='viridis')

# Add colorbar and set properties
cbar = fig.colorbar(contour, ax=ax, label='Gaussian Value')
cbar.set_label('Gaussian Value', fontsize=cbar_fs)
cbar.ax.tick_params(labelsize=axis_num_fs, which='major', length=7, direction="in")
cbar.ax.tick_params(which='minor', length=4, direction="in")

# Set title and labels
ax.set_title('2D Gaussian Distribution', fontsize=title_fs)
ax.set_xlabel('X-axis', fontsize=axis_label_fs)
ax.set_ylabel('Y-axis', fontsize=axis_label_fs)

# Adjust ticks
ax.minorticks_on()
ax.tick_params(axis="x", which="major", direction="in", bottom=True, top=True, length=7, labelsize=axis_num_fs)
ax.tick_params(axis="y", which="major", direction="in", left=True, right=True, length=7, labelsize=axis_num_fs)

plt.show()
```

```python
# Plotting the heatmap
fig, ax = plt.subplots(figsize=(13, 10))

# Plotting main heatmap
im = ax.pcolormesh(A,
                   B,
                   D,
                   cmap='Greens')

# Add colorbar and set properties
cbar = fig.colorbar(im, ax=ax, label='Gaussian Value')
cbar.set_label('Gaussian Value', fontsize=cbar_fs)
cbar.ax.tick_params(labelsize=axis_num_fs, which='major', length=7, direction="in")
cbar.ax.tick_params(which='minor', length=4, direction="in")


# Set title and labels
ax.set_title('2D Gaussian Distribution', fontsize=title_fs)
ax.set_xlabel('X-axis', fontsize=axis_label_fs)
ax.set_ylabel('Y-axis', fontsize=axis_label_fs)

# Adjust ticks
ax.minorticks_on()
ax.tick_params(axis="x", which="major", direction="in", bottom=True, top=True, length=7, labelsize=axis_num_fs)
ax.tick_params(axis="y", which="major", direction="in", left=True, right=True, length=7, labelsize=axis_num_fs)

plt.show()
```

Now lets try with 3D

```python
# Dictionary to store results
results_3d = []

c_const = 1

for a in a_values: 
    for b in b_values: 
        for c in c_values: 
            
            d = d_function(a, b, c)

            results_3d.append((a, b, c, d))
```

```python
# Convert results into structured 2D arrays
a_list, b_list, c_list, d_list = zip(*results_3d)  # Unpack into separate lists

# Reshape into 2D grids (len(a_values) rows, len(b_values) columns)
# Reshape into 3D grids
AA = np.array(a_list).reshape(len(a_values), len(b_values), len(c_values))
BB = np.array(b_list).reshape(len(a_values), len(b_values), len(c_values))
CC = np.array(c_list).reshape(len(a_values), len(b_values), len(c_values))
DD = np.array(d_list).reshape(len(a_values), len(b_values), len(c_values))
```

```python
# Plotting the result
fig, ax = plt.subplots(2, 2, figsize=(16, 18))

# Ensure c_values has at least 4 elements
num_plots = min(4, len(c_values))  

# Loop over the first `num_plots` values of c
for i, c_fixed_idx in enumerate(range(num_plots)):  

    A = AA[:, :, c_fixed_idx]  # Slice at fixed c
    B = BB[:, :, c_fixed_idx]
    D = DD[:, :, c_fixed_idx]
    
    row = i // 2  # Row index (0 or 1)
    col = i % 2   # Column index (0 or 1)
    
    # Create the contour plot
    contour = ax[row, col].contourf(A, B, D, 50, cmap='viridis')
    
    # Add colorbar
    cbar = fig.colorbar(contour, ax=ax[row, col], orientation='vertical')
    cbar.set_label('D', fontsize=cbar_fs)
    cbar.ax.tick_params(labelsize=axis_num_fs, which='major', length=7, direction="in")
    cbar.ax.tick_params(which='minor', length=4, direction="in")
    
    # Set title with actual c-value
    ax[row, col].set_title(rf'C = {c_values[c_fixed_idx]}', fontsize=title_fs)  

    # Labels
    ax[row, col].set_xlabel('A', fontsize=axis_label_fs)
    ax[row, col].set_ylabel('B', fontsize=axis_label_fs)

    # Ticks
    ax[row, col].minorticks_on()
    ax[row, col].tick_params(axis="x", which="major", direction="in", bottom=True, top=True, length=7, labelsize=axis_num_fs)
    ax[row, col].tick_params(axis="y", which="major", direction="in", left=True, right=True, length=7, labelsize=axis_num_fs)

plt.tight_layout()
plt.show()
```

```python
# Loop over the first `num_plots` values of B
for i, b_fixed_idx in enumerate(range(num_plots)):  

    A = AA[:, b_fixed_idx, :]  # Slice at fixed B
    C = CC[:, b_fixed_idx, :]
    D = DD[:, b_fixed_idx, :]  # Adjust for new axis arrangement
    
    row = i // 2  # Row index (0 or 1)
    col = i % 2   # Column index (0 or 1)
    
    # Create the contour plot
    contour = ax[row, col].contourf(A, C, D, 50, cmap='viridis')
    
    # Add colorbar
    cbar = fig.colorbar(contour, ax=ax[row, col], orientation='vertical')
    cbar.set_label('D', fontsize=cbar_fs)
    cbar.ax.tick_params(labelsize=axis_num_fs, which='major', length=7, direction="in")
    cbar.ax.tick_params(which='minor', length=4, direction="in")
    
    # Set title with actual B-value
    ax[row, col].set_title(rf'B = {BB[0, b_fixed_idx, 0]}', fontsize=title_fs)  

    # Labels
    ax[row, col].set_xlabel('A', fontsize=axis_label_fs)
    ax[row, col].set_ylabel('C', fontsize=axis_label_fs)

    # Ticks
    ax[row, col].minorticks_on()
    ax[row, col].tick_params(axis="x", which="major", direction="in", bottom=True, top=True, length=7, labelsize=axis_num_fs)
    ax[row, col].tick_params(axis="y", which="major", direction="in", left=True, right=True, length=7, labelsize=axis_num_fs)

plt.tight_layout()
plt.show()
```

```python
# Plotting the result
fig, ax = plt.subplots(2, 2, figsize=(16, 18))

# Assuming phi_values is a list or array with at least 4 elements
for i, c_fixed_index in enumerate(c_values[:4]):  # Loop over the first 4 values of phi
    
    A = AA[:, :, c_fixed_idx]
    B = BB[:, :, c_fixed_idx]
    D = DD[:, :, c_fixed_idx]
    
    # Calculate the row and column position
    row = i // 2  # Determine the row (0 or 1)
    col = i % 2   # Determine the column (0 or 1)
    
    # Create the contour plot
    contour = ax[row, col].contourf(A, B, D, 50, cmap='viridis')
    
    # Add colorbar to each subplot and set properties
    cbar = fig.colorbar(contour, ax=ax[row, col], orientation='vertical')
    cbar.set_label('D', fontsize=cbar_fs)
    cbar.ax.tick_params(labelsize=axis_num_fs, which='major', length=7, direction="in")
    cbar.ax.tick_params(which='minor', length=4, direction="in")
    
    # Set the title
    ax[row, col].set_title(rf'C = {c_fixed_index}', fontsize=title_fs)  # Correct formatting with f-string
    
    # Set the x and y labels
    ax[row, col].set_xlabel('A', fontsize=axis_label_fs)
    ax[row, col].set_ylabel('B', fontsize=axis_label_fs)

    # Set size of ticks and labels
    ax[row, col].minorticks_on()
    ax[row, col].tick_params(axis="x", which="major", direction="in", bottom=True, top=True, length=7, labelsize=axis_num_fs)
    ax[row, col].tick_params(axis="y", which="major", direction="in", left=True, right=True, length=7, labelsize=axis_num_fs)
```

## 3D Plot

```python
from scipy.stats import multivariate_normal
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Create a bigger figure
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surf = ax.plot_surface(AA, BB, DD, cmap='viridis', linewidth=0)

# Add colorbar and set properties
cbar = fig.colorbar(surf, ax=ax, label='Gaussian Value')
cbar.set_label('Gaussian Value', fontsize=cbar_fs)
cbar.ax.tick_params(labelsize=axis_num_fs, which='major', length=7, direction="in")
cbar.ax.tick_params(which='minor', length=4, direction="in")

# Set title and labels with font size
ax.set_title('2D Gaussian Distribution', fontsize=title_fs)
ax.set_xlabel('X-axis', fontsize=axis_label_fs)
ax.set_ylabel('Y-axis', fontsize=axis_label_fs)
ax.set_zlabel('Z-axis', fontsize=axis_label_fs)

# Adjust ticks
ax.tick_params(axis="x", which="major", direction="in", bottom=True, top=True, length=7, labelsize=axis_num_fs)
ax.tick_params(axis="y", which="major", direction="in", left=True, right=True, length=7, labelsize=axis_num_fs)
ax.tick_params(axis="z", which="major", direction="in", length=7, labelsize=axis_num_fs)

plt.show()
```

```python
# Select a slice (e.g., along the third axis, `c_values` axis)
# You can change this index based on which slice of `c_values` you'd like to plot
slice_index = len(c_values) // 2  # Midway through the `c_values` axis
AA_slice = AA[:, :, slice_index]
BB_slice = BB[:, :, slice_index]
DD_slice = DD[:, :, slice_index]

# Create a bigger figure
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface using the selected slice of AA, BB, and DD
surf = ax.plot_surface(AA_slice, BB_slice, DD_slice, cmap='viridis', linewidth=0)

# Add colorbar and set properties
cbar = fig.colorbar(surf, ax=ax, label='D Value')
cbar.set_label('D Value', fontsize=cbar_fs)
cbar.ax.tick_params(labelsize=axis_num_fs, which='major', length=7, direction="in")
cbar.ax.tick_params(which='minor', length=4, direction="in")

# Set title and labels with font size
ax.set_title('3D Surface Plot of AA, BB, and DD', fontsize=title_fs)
ax.set_xlabel('A values (X-axis)', fontsize=axis_label_fs)
ax.set_ylabel('B values (Y-axis)', fontsize=axis_label_fs)
ax.set_zlabel('D values (Z-axis)', fontsize=axis_label_fs)

# Adjust ticks
ax.tick_params(axis="x", which="major", direction="in", bottom=True, top=True, length=7, labelsize=axis_num_fs)
ax.tick_params(axis="y", which="major", direction="in", left=True, right=True, length=7, labelsize=axis_num_fs)
ax.tick_params(axis="z", which="major", direction="in", length=7, labelsize=axis_num_fs)

plt.show()
```
