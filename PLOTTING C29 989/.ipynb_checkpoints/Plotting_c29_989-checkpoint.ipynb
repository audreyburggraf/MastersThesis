## Calculate Polarization Angle

$$
\theta = \frac{1}{2} \arctan\left(\frac{U}{Q}\right)  \text{   radians}
$$

I got this equation from [this paper](https://www.sciencedirect.com/science/article/pii/S0022407314003744#:~:text=The%20angle%20of%20polarization%20%CF%87,with%20the%20angle%20of%20polarization.)
- this angle is measured using the cartesian coordinates so we do not need to convert it to plot it

````python
calculated_polarization_angle_rad = 0.5 * np.arctan2(StokesU_data_2d_Jy, StokesQ_data_2d_Jy)
````

### Make sure that my calculations are correct

````python
# Calculate the difference
Polarization_Angle_Difference = calculated_polarization_angle_rad - PolarizationAngle_data_2d_rad

# Perform calculations
sum_difference = np.nansum(Polarization_Angle_Difference)
min_difference = np.nanmin(Polarization_Angle_Difference)
max_difference = np.nanmax(Polarization_Angle_Difference)

# Print the results
print("The sum of differences is:", sum_difference)
print("The minimum difference is:", min_difference)
print("The maximum difference is:", max_difference)
````

````python
test_value = 2

# Count how many elements are greater than test_value
num_max_values = np.sum(Polarization_Angle_Difference > test_value)

# Print the result
print(f"The number of values with a difference above {test_value} is:", num_max_values)
````

````python
# Find the indices where the difference equals the maximum value
indices_max = np.where(Polarization_Angle_Difference == max_difference)

# Extract the corresponding polarization angle values at those indices
polarization_angle_rad_max = calculated_polarization_angle_rad[indices_max]
PolarizationAngle_data_2d_rad_max = PolarizationAngle_data_2d_rad[indices_max]

# Extract Stokes Q and U values at those indices
StokesQ_max = StokesQ_data_2d_mJy[indices_max]
StokesU_max = StokesU_data_2d_mJy[indices_max]

# Print the results
print("Calculated Polarization Angle Rad (where max difference):", polarization_angle_rad_max)
print("From Data Polarization Angle Data (where max difference):", PolarizationAngle_data_2d_rad_max)
print("Stokes Q (where max difference):", StokesQ_max)
print("Stokes U (where max difference):", StokesU_max)
````
