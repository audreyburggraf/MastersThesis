```python
testing_uniform_vector_data = []

# Loop over values in x and y
for x in range(0, nx, step):
    for y in range(0, ny, step):
        # Check if the conditions are met
        if (StokesI_data_2d_mJy[y, x] / StokesIerr_data_2d_mJy[y, x] > 3 and 
            calculated_polarized_intensity[y, x] / PolarizedIntensity_err_data_2d_mJy[y, x] > 3 and 
            PolarizationAngle_err_data_2d_deg[y, x] < 10):
            
            angle_rad = PA_grid_100Uniform[y, x]
            
            # Compute the vector components
            dx = vector_length_pix_const * np.cos(angle_rad + np.pi/2)
            dy = vector_length_pix_const * np.sin(angle_rad + np.pi/2)
            
            # Store vector data in a row
            testing_uniform_vector_data.append([x - dx / 2, x + dx / 2, y - dy / 2, y + dy / 2])
# --------------------------------------------------------------------------------------------------              
```

I would expect the vectors to be aligned with the minor axis visually after we changed it to cartesian

```python
fig, ax = create_base_plot(StokesI_wcs, StokesI_stretched, soft_colormap_v2, 
                           normalized_cbar_ticks, StokesI_unstretched_cbar_ticks, 
                           xmin, xmax, ymin, ymax, reference_length_pix, reference_length_AU,
                           text_fs, axis_label_fs, axis_num_fs, cbar_fs,
                           BMAJ_pix, BMIN_pix, BPA_deg_cartesian, 
                           max_length_pix, reference_fraction)

for row in testing_uniform_vector_data:
    ax.plot([row[0], row[1]], [row[2], row[3]], color='black', lw = 3)
```

## PA_grid_100Azimuthal

At this point PA_grid_100Azimuthal should have azimuthal angles in astronomy and not cartesian

```python
testing_azimuthal_vector_data = []

# Loop over values in x and y
for x in range(0, nx, step):
    for y in range(0, ny, step):
        # Check if the conditions are met
        if (StokesI_data_2d_mJy[y, x] / StokesIerr_data_2d_mJy[y, x] > 3 and 
            calculated_polarized_intensity[y, x] / PolarizedIntensity_err_data_2d_mJy[y, x] > 3 and 
            PolarizationAngle_err_data_2d_deg[y, x] < 10):
            
            angle_rad = PA_grid_100Azimuthal[y, x]
            
            # Compute the vector components
            dx = vector_length_pix_const * np.cos(angle_rad + np.pi/2)
            dy = vector_length_pix_const * np.sin(angle_rad + np.pi/2)
            
            # Store vector data in a row
            testing_azimuthal_vector_data.append([x - dx / 2, x + dx / 2, y - dy / 2, y + dy / 2])
# --------------------------------------------------------------------------------------------------              
```

At this point the plot should visually be azimuthal/circular as it is not comverted to cartesian
