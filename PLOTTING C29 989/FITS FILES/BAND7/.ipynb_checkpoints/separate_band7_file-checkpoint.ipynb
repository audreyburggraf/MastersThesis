```python
from astropy.io import fits

hdul = fits.open('member.uid___A001_X133d_X15c9.IRS63_sci.spw13_15_17_19.mfs.IQUV.manual.pbcor.fits')
data = hdul[0].data  # Typically shape is (4, y, x) for I,Q,U,V

header = hdul[0].header
# Separate the Stokes components
I = data[0, :, :, :]
Q = data[1, :, :, :]
U = data[2, :, :, :]

# You can save each one using fits.writeto if needed
from astropy.io.fits import writeto

writeto('IRS63_StokesI_uncalibrated_BAND7.fits', I, header, overwrite=True)
writeto('IRS63_StokesQ_uncalibrated_BAND7.fits', Q, header, overwrite=True)
writeto('IRS63_StokesU_uncalibrated_BAND7.fits', U, header, overwrite=True)
```
