## 2D Flat Topped Gaussian

```python
# Store the computed values
gaussian_2d_flat = {}
gaussian_slices_x = {}
gaussian_slices_y = {}

phi_values = [2, 3, 4, 5]  # Increasing phi makes it flatter

# Compute and store the data
for phi in phi_values:
    gaussian_2d = calculate_gaussian_2d_flat(x_values, y_values, 2, 3, mean_x, mean_y, phi)
    
    gaussian_2d_flat[phi]  = gaussian_2d.copy()  # Store the slice along x
    gaussian_slices_x[phi] = gaussian_2d[50, :].copy()  # Store the slice along x
    gaussian_slices_y[phi] = gaussian_2d[:, 50].copy()  # Store the slice along y
```

```python
# Plot the stored data
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

for phi in phi_values:
    ax[0].plot(x_values, gaussian_slices_x[phi], ls='-', label=fr'$\phi={phi}$')
    ax[1].plot(y_values, gaussian_slices_y[phi], ls='-', label=fr'$\phi={phi}$')

# Formatting
ax[0].set_title('2D Flat Gaussian in X', fontsize=title_fs)
ax[0].set_xlabel("x-values", fontsize=axis_label_fs)
ax[0].set_ylabel("Function Value", fontsize=axis_label_fs)
ax[0].legend()

ax[1].set_title('2D Flat Gaussian in Y', fontsize=title_fs)
ax[1].set_xlabel("y-values", fontsize=axis_label_fs)
ax[1].set_ylabel("Function Value", fontsize=axis_label_fs)
ax[1].legend()

plt.tight_layout()
plt.show()
```

```python
# Create subplots
fig, ax = plt.subplots(2, 2, figsize=(18, 14))  
ax = ax.flatten()  # Flatten for easy looping

# Loop through stored data and plot
for i, phi in enumerate(phi_values):
    # Retrieve stored 2D Gaussian
    gaussian_2d = gaussian_2d_flat[phi]

    # Contour plot in each subplot
    contour = ax[i].contourf(x_values, y_values, gaussian_2d, 50, cmap='viridis')

    # Add individual colorbar for each subplot
    cbar = fig.colorbar(contour, ax=ax[i], orientation='vertical', fraction=0.05, pad=0.05)
    cbar.set_label('Gaussian Value', fontsize=cbar_fs)
    cbar.ax.tick_params(labelsize=axis_num_fs, which='major', length=7, direction="in")
    cbar.ax.tick_params(which='minor', length=4, direction="in")

    # Set title and labels
    ax[i].set_title(fr'2D Gaussian ($\phi={phi}$)', fontsize=title_fs)
    ax[i].set_xlabel('X-axis', fontsize=axis_label_fs)
    ax[i].set_ylabel('Y-axis', fontsize=axis_label_fs)
    
    # Adjust ticks
    ax[i].minorticks_on()
    ax[i].tick_params(axis="x", which="major", direction="in", bottom=True, top=True, length=7, labelsize=axis_num_fs)
    ax[i].tick_params(axis="y", which="major", direction="in", left=True, right=True, length=7, labelsize=axis_num_fs)

plt.tight_layout()
plt.show()
```

## 2D Flat Gaussian Along Tilted Axes

```python
# Store the computed values
gaussian_2d_flat_tilted = {}
gaussian_slices_x_tilted = {}
gaussian_slices_y_tilted = {}
X_rot_arr = {}
Y_rot_arr = {}

phi_values = [2, 3, 4, 5]  # Increasing phi makes it flatter
theta_rad_values = np.radians([45, 90, 135, 180])  # Corrected syntax

# Compute and store the data
for phi, theta_rad in zip(phi_values, theta_rad_values):
    gaussian_2d_tilted, X_rot, Y_rot = calculate_gaussian_2d_flat_tilted(x_values, y_values, 2, 3, mean_x, mean_y, phi, theta_rad)

    # Store using (phi, theta) as the key
    gaussian_2d_flat_tilted[(phi, theta_rad)] = gaussian_2d_tilted.copy()
    
    X_rot_arr[(phi, theta_rad)] = X_rot.copy()
    Y_rot_arr[(phi, theta_rad)] = Y_rot.copy()
    
    gaussian_slices_x_tilted[(phi, theta_rad)] = gaussian_2d_tilted[50, :].copy()
    gaussian_slices_y_tilted[(phi, theta_rad)] = gaussian_2d_tilted[:, 50].copy()
```

```python
# Plot the stored data
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

for phi, theta_rad in zip(phi_values, theta_rad_values):
    
    
    # Extract the 1D slice for x and y, ensuring they match the dimensions of x_values and y_values
    slice_x = gaussian_slices_x_tilted[(phi, theta_rad)]
    slice_y = gaussian_slices_y_tilted[(phi, theta_rad)]
    
    x_values_rot = X_rot_arr[(phi, theta_rad)]
    y_values_rot = Y_rot_arr[(phi, theta_rad)]

    ax[0].plot(x_values_rot[50, :], slice_x, ls='-', label=fr'$\phi={phi}, \theta={theta_rad * 180/np.pi:.0f}^\circ$')
    ax[1].plot(y_values_rot[:, 50], slice_y, ls='-', label=fr'$\phi={phi}, \theta={theta_rad * 180/np.pi:.0f}^\circ$')


# Formatting
ax[0].set_title('2D Flat Gaussian in X', fontsize=title_fs)
ax[0].set_xlabel("x-values", fontsize=axis_label_fs)
ax[0].set_ylabel("Function Value", fontsize=axis_label_fs)
ax[0].legend()

ax[1].set_title('2D Flat Gaussian in Y', fontsize=title_fs)
ax[1].set_xlabel("y-values", fontsize=axis_label_fs)
ax[1].set_ylabel("Function Value", fontsize=axis_label_fs)
ax[1].legend()

plt.tight_layout()
plt.show()
```

```python
# Create subplots: 2D Gaussian contours (2x2 grid)
fig, ax = plt.subplots(2, 2, figsize=(18, 14))  
ax = ax.flatten()


i = 0 

# Plot the 2D Gaussian grids
for phi, theta_rad in zip(phi_values, theta_rad_values):
    
    gaussian_2d = gaussian_2d_flat_tilted[(phi, theta_rad)]
    
    
    x_values_rot = X_rot_arr[(phi, theta_rad)][50, :]
    y_values_rot = Y_rot_arr[(phi, theta_rad)][:, 50]
    
    contour = ax[i].contourf(x_values, y_values, gaussian_2d, 50, cmap='viridis')
    cbar = fig.colorbar(contour, ax=ax[i], orientation='vertical', fraction=0.05, pad=0.05)
    cbar.set_label('Gaussian Value', fontsize=cbar_fs)
    
    ax[i].set_title(fr'2D Gaussian Flat Tilted ($\phi={phi}, \theta={theta_rad * 180/np.pi:.0f}^\circ$)', fontsize=title_fs)
    ax[i].set_xlabel('X-axis', fontsize=axis_label_fs)
    ax[i].set_ylabel('Y-axis', fontsize=axis_label_fs)
    ax[i].minorticks_on()
    ax[i].tick_params(axis="x", which="major", direction="in", bottom=True, top=True, length=7, labelsize=axis_num_fs)
    ax[i].tick_params(axis="y", which="major", direction="in", left=True, right=True, length=7, labelsize=axis_num_fs)
    
    i = i + 1

plt.tight_layout()
plt.show()
```
